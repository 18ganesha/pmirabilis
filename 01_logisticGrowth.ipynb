```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic growth function
def logistic_growth(t, r, K, rho_0):
    return rho_0 / (rho_0/rho_c+(1-rho_0/rho_c)*np.exp(-G*t))

# Parameters
G = 0.5  # Growth rate
rho_c = 1  # Carrying capacity
init_rho = [0.001, 0.01, .1, .2, .5, .7, 1, 1.2, 2, 10]  # Different initial densities

t = np.linspace(0, 20, 500)  # Time points for the plot

# Plot the logistic growth for each initial density
plt.figure(figsize=(10, 6))

for rho_0 in init_rho:
    rho = logistic_growth(t, G, rho_c, rho_0)
    plt.plot(t, np.log(rho), label=f"rho_0 = {rho_0}")

# Customize the plot
plt.title("Logistic Growth (analytical)", fontsize=14)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Population Density", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```
